## Notação

1. **Notação  $ ( O ) $   (Big O):** Descreve um limite superior no crescimento de uma função. Se $ f(n) = O(g(n)) $, isso significa que existe uma constante  $ C > 0 $ tal que $ f(n) \leq C \cdot g(n) $ para todos os $ n \geq n_0 $ suficientemente grandes.

2. **Notação \( $\Theta$ \) (Big Theta):** Descreve uma função que tem tanto um limite superior quanto um limite inferior assintoticamente apertados. Se $  f(n) = \Theta(g(n))  $, isso significa que existem constantes $ C_1, C_2 > 0 $ tais que $ C_1 \cdot g(n) \leq f(n) \leq C_2 \cdot g(n) $ para todos os $ n \geq n_0 $ suficientemente grandes.

3. **Notação $( \Omega )$ (Big Omega):** Descreve um limite inferior no crescimento de uma função. Se $ f(n) = \Omega(g(n)) $, isso significa que existe uma constante $ C > 0 $ tal que $ f(n) \geq C \cdot g(n) $ para todos os $ n \geq n_0 $ suficientemente grandes.

4. **Notação $( o )$ (Little o):** Descreve um limite superior estrito. Se $ f(n) = o(g(n)) $, isso significa que $\lim_{n \to \infty} \frac{f(n)}{g(n)} = 0 $. Em outras palavras, $ f(n) $ cresce estritamente mais devagar do que $ g(n) $.

5. **Notação $ ( \omega ) $ (Little Omega):** Descreve um limite inferior estrito. Se $  f(n) = \omega(g(n)) $, isso significa que $ \lim_{n \to \infty} \frac{f(n)}{g(n)} = \infty $. Em outras palavras, \( f(n) \) cresce estritamente mais rápido do que $ g(n) $.


## Teorema Mestre
O Teorema Mestre é uma ferramenta essencial na análise de algoritmos, particularmente para algoritmos de divisão e conquista. Ele fornece uma maneira rápida e fácil de determinar a complexidade de tempo de uma equação de recorrência com uma forma específica.

### Forma da Recorrência
O Teorema Mestre aplica-se a recorrências da seguinte forma:

$ T(n) = a \cdot T\left(\frac{n}{b}\right) + f(n) $

onde:
- $ a $ é o número de subproblemas na recursão.
- $ b > 1 $ é o fator pelo qual o tamanho do problema é dividido.
- $ f(n) $ é o custo do trabalho feito fora das chamadas recursivas (como a combinação de subproblemas).

### Três Casos
O teorema divide-se em três casos, que determinam a complexidade da recorrência:

1. **Caso 1:** Se $ f(n) = O(n^{\log_b{a} - \epsilon}) $ para algum $ \epsilon > 0 $, então a complexidade é $ \Theta(n^{\log_b{a}}) $.
2. **Caso 2:** Se $ f(n) = \Theta(n^{\log_b{a}}) $, então a complexidade é $ \Theta(n^{\log_b{a}} \log n) $.
3. **Caso 3:** Se $ f(n) = \Omega(n^{\log_b{a} + \epsilon}) $ para algum $ \epsilon > 0 $, e se $ a f\left(\frac{n}{b}\right) \leq kf(n) $ para algum $ k < 1 $ e $ n \geq n_0 $ suficientemente grande, então a complexidade é $ \Theta(f(n)) $.

### Limitações
O Teorema Mestre não se aplica a todas as recorrências. Deve ser usado apenas quando a recorrência se encaixa na forma acima e satisfaz um dos três casos.

### Utilização
O Teorema Mestre é comumente usado para analisar algoritmos de divisão e conquista, como o algoritmo de ordenação merge sort e o algoritmo de busca binária, fornecendo uma maneira eficiente de determinar a complexidade de tempo sem resolver a recorrência manualmente.

### Questão 01
Dados:
1. Para um 10 mil documentos demora 1s. Que dá $ 10^4 $ operações.
2. Para 100 milhões são $ 10^8 $ operações.

Resolução:

Podemos calcular o fator de crescimento do número de operações como a razão dos logaritmos das duas quantidades:

$$ \frac{\log 10^8}{\log 10^4} = 2 $$

Este fator representa o quanto o número de operações cresceu. Para encontrar o tempo correspondente, multiplicamos esse fator pelo tempo original:

$$ 2 * 1 = 2s $$


### Questão 2: Complexidade $O(n^2)$
Um cientista da computação executa um algoritmo de ordenação com complexidade quadrática em um conjunto de 1000 números e leva 2 segundos. Quantos segundos seriam necessários para ordenar 5000 números, mantendo o mesmo custo unitário de operação?

### Resolução:

Dados:
1. Para um conjunto de 1000 elementos demora 1s. Que dá $ (10^3)^2 = 10^6 $ operações.
2. Para 5000 mil são $ (5*10^3)^2 = 5^2 * 10^6 $ operações.

Resolvendo:

$$ \frac{5^2 * 10^6}{10^6} = 5^2 $$

Por fim, temos que o tempo de execução é: $25*2s = 50s$.


### Questão 3: Complexidade $O(2^n)$
Um pesquisador testa um algoritmo recursivo que resolve um problema NP-completo em uma instância de tamanho 10 e leva 1 segundo. Quantos segundos seriam necessários para resolver uma instância de tamanho 20, considerando que o custo cresce exponencialmente com o tamanho da entrada?

### Resolução:

Dados:
1. Para um conjunto de 10 elementos demora 1000s. Que dá $ 2^{10} $ operações.
2. Para 20 são $ 2^{20} $ operações.

Resolvendo:

$$ \frac{2^{20}}{2^{10}} = 2^{10} $$

Por fim, temos que o tempo de execução é: $2^{10}*1s = 1024s$.

### Resolução:

````python
code = """
#include <stdio.h>

int bubbleSort(int* A, int n){

  for(int i=n-1; i > 0; i--){

    for(int j=0; j < i; j++){

      if(A[j] < A[j+1]){
        int tmp = A[j];
        A[j] = A[j+1];
        A[j+1] = tmp;
      }

    }

  }

}

void printVector(int* A, int n){

    printf("{ ");
    for(int i=0; i < n; i++){
      printf("%d, ", A[i]);
    }
    printf(" }\\n");
}

int main(){
  int arr[] = {6, 3, 0, 5};

  printVector(arr,4);
  bubbleSort(arr,4);
  printVector(arr,4);

}

"""

with open("bubble.c", "w") as f:
  f.write(code)
````

2. Para a função interna temos: $\sum _{i=0}^{n-i-1}\left(1\right)$

### Calculando o mais interno temos:

$$
\sum _{i=a}^{n}\left(1\right) = (n-a)+1
$$

logo,
$$
\sum _{j=0}^{n-i-1}\left(1\right) = (n-i-1-0)+1 = n-i
$$
Agora revendo o somatorio mais interno temos:

$$
\sum _{i=1}^{n-1}\left(n-i\right)
$$

Isso é o mesmo que:

$$
\sum _{i=1}^{n-1}\left(n\right) - \sum _{i=1}^{n-1}\left(i\right)
$$

Que é o mesmo que:
$$
n*\sum _{i=1}^{n-1}\left(1\right) - \sum _{i=1}^{n-1}\left(i\right)
$$

Calculando $n*\sum _{i=1}^{n-1}\left(1\right)$ temos:
$$
n*[(n-1-1)+1] = n*[(n-2)+1] = n*n-1
$$
Por fim, essa parte resulta em: $ n^2 - n $

Agora Calculando a segunda parte da equanção temos:
$$
\sum _{i=1}^{n}\left(i\right) = \frac{n(n+1)}{2}
$$

Logo,
$$
\sum _{i=1}^{n-1}\left(i\right) = \frac{n-1(n-1+1)}{2} = \frac{n-1(n)}{2} = \frac{n^2-n}{2}
$$

$$
 n^2 - n -\frac{n^2-n}{2} = \frac{2n^2 - 2n - n^2 - n}{2} = \frac{n^2-3
 n}{2}
$$

Portanto a equação fechado do bubble sort é $ \frac{n^2-3

````python
code = """
#include <stdio.h>

int insertionSort(int* A, int n){

  int pivo;
  int i;
  for(i=1; i < n; i++){

    pivo = A[i];

    int j = 0;
    while (j < i && A[i-1-j] > pivo)
    {
        A[i-j] = A[i-1-j];
        j++;
    }

    A[i-j] = pivo;


  }

}


void printVector(int* A, int n){

    printf("{ ");
    for(int i=0; i < n; i++){
      printf("%d, ", A[i]);
    }
    printf(" }\\n");
}

int main(){
  int arr[] = {6, 3, 0, 5};

  printVector(arr,4);
  insertionSort(arr,4);
  printVector(arr,4);

}

"""

with open("insertion_sort.c", "w") as f:
  f.write(code)
````
